# 1. 전처리

```python
# 데이터 확인

raw_data = pd.read_excel("농업과학원 전달_통합본_20221123_20230306.xlsx", sheet_name=2)
raw_data.info()
```

`-` 고찰
1) 원본데이터(HEX)는 필요 없어보임 => 삭제
2) 데이터구분이 필요하다 => raw_dataA, raw_dataB 이런식으로 데이터 split 필요
3) 이벤트발생일을 시계열 데이터로 처리 => datetime으로 변환후 split

```python
#1. 원본데이터(HEX) 삭제
raw_data.drop(columns=['원본데이터(HEX)', '원본데이터(HEX).1'], inplace=True)
```

```python
#2. 데이터 구분
raw_dataA = raw_data.iloc[:,:17]
raw_dataB = raw_data.iloc[:,17:]
```

```python
#3. datetime 변경
raw_dataA["이벤트발생일"]=pd.to_datetime(raw_dataA["이벤트발생일"], format='%Y-%m-%d %H:%M:%S') #inplace 파라미터 없냐 귀찮네
raw_dataB["이벤트발생일.1"]=pd.to_datetime(raw_dataB["이벤트발생일.1"], format='%Y-%m-%d %H:%M:%S')

raw_dataA['년']=raw_dataA['이벤트발생일'].dt.year
raw_dataA['월']=raw_dataA['이벤트발생일'].dt.month
raw_dataA['일']=raw_dataA['이벤트발생일'].dt.day
raw_dataA['시']=raw_dataA['이벤트발생일'].dt.hour
raw_dataA['분']=raw_dataA['이벤트발생일'].dt.minute
raw_dataA['초']=raw_dataA['이벤트발생일'].dt.second   #시발 존나 귀찮은데? 클래스로 하는 방법 없을까?

raw_dataB['년']=raw_dataB['이벤트발생일.1'].dt.year
raw_dataB['월']=raw_dataB['이벤트발생일.1'].dt.month
raw_dataB['일']=raw_dataB['이벤트발생일.1'].dt.day
raw_dataB['시']=raw_dataB['이벤트발생일.1'].dt.hour
raw_dataB['분']=raw_dataB['이벤트발생일.1'].dt.minute
raw_dataB['초']=raw_dataB['이벤트발생일.1'].dt.second
```

`-` 공부가 덜되었으니 복붙작업 하겠음..

```python
# 데이터 확인

raw_data2 = pd.read_excel("농업과학원 전달_통합본_20221123_20230306.xlsx", sheet_name=3)
raw_data2.info()
```

## 병합

```python
data1=pd.concat([raw_dataA,raw_data_A], ignore_index=True)
data2=pd.concat([raw_dataB,raw_data_B], ignore_index=True)
```
